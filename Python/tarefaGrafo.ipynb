```python
""" 
Aluno: Akira Saito
Cód.: 831474
Prof.: Rodrigo de Oliveira Plotze
Diciplina: Teoria da Computação 
"""

# Importa uma biblioteca para trabalhar com grafos 
import networkx as nx # Library for displaying graphs.

# Cria um grafo vazio
G = nx.Graph()

 #criar os vértices do grafo
G.add_node(0)
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)
G.add_node(6)
G.add_node(7)

#definir as arestas do grafo
G.add_edge(0,2)
G.add_edge(0,4)
G.add_edge(0,6)
G.add_edge(1,3)
G.add_edge(1,5)
G.add_edge(2,0)
G.add_edge(2,6)
G.add_edge(2,7)
G.add_edge(3,1)
G.add_edge(3,6)
G.add_edge(3,7)
G.add_edge(4,1)
G.add_edge(4,5)
G.add_edge(4,6)
G.add_edge(4,7)
G.add_edge(5,1)
G.add_edge(5,4)
G.add_edge(5,7)
G.add_edge(6,1)
G.add_edge(6,2)
G.add_edge(6,3)
G.add_edge(6,4)
G.add_edge(7,2)
G.add_edge(7,3)
G.add_edge(7,4)
G.add_edge(7,5)

#armazenar em uma matriz adjacente
A = nx.adjacency_matrix(G)
print(A.todense())  #retornar matriz adjacente

print("\n--------------------------------------------\n")

#perguntar ao usuario qual o caminho que ele quer saber 
V1 = int(input("\nInforme o vértice inicial: "))
V2 = int(input("\nInforme o vértice final: "))

#retorna os vertices conectados pelas arestas n-1
P = nx.path_graph(G)

#retorna o caminho mais curto até o destino 
print(nx.shortest_path(P, source = V1, target = V2))

print("\n--------------------------------------------\n")

#visualizar o grafo
nx.draw(G,with_labels=True)
```
