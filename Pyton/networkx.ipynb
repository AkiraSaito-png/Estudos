```python
# Grafo Não-Direcionado

#importação do pacote NetworkX
import networkx as nx

#criar um novo Grafo NÃO-DIRECIONADO
G = nx.Graph()

#criar os vértices do grafo
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_nodes_from(["F","G","H"])

#definir as arestas do grafo
G.add_edge("A","B")
G.add_edge("A","C")
G.add_edge("B","C")
G.add_edges_from([("D","A"),("D","B"),("D","C"),("F","G"),("G","H"),("G","D")])

#visualizar o grafo
nx.draw(G,with_labels=True)
```

```python
# Grafo Direcionado

import networkx as nx

#criar um novo Grafo DIRECIONADO
G = nx.DiGraph()

G.add_node("A")
G.add_nodes_from(["B","C","D"])

G.add_edge("A","B")
G.add_edge("A","C")
G.add_edge("B","C")
G.add_edges_from([("D","A"),("D","B"),("D","C")])

nx.draw(G,with_labels=True)
```
