```python
# Crie uma aplicação capaz de representar um conjunto de vértices V={1,2,3,4,5,6} e um conjunto de arestas A={(1,3), (1,4), (1,5), (2,3), (2,4), (2,5), (3,5), (4,5)}, utilizando um grafo não-direcionado.

#importar do pacote NetWorkx
import networkx as nx

G = nx.Graph()

#criar os vertices do grafo
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)
G.add_node(6)

#definir as arestas do grafo
G.add_edge(1,3)
G.add_edge(1,4)
G.add_edge(1,5)
G.add_edge(2,3)
G.add_edge(2,4)
G.add_edge(2,5)
G.add_edge(3,5)
G.add_edge(4,5)

nx.draw(G, with_labels = True)
```

```python
# Crie uma aplicação capaz de representar um grafo não-direcionado com as seguintes características: 1 vértice com grau=3, 1 vértice com grau=2, 1 vértice com grau=1 e 1 vértice com grau=0.

import networkx as nx

G = nx.Graph()

G.add_node(3)
G.add_node(2)
G.add_node(1)
G.add_node(0)

nx.draw(G, with_labels = True)
```

```python
# Crie uma aplicação capaz de representar um grafo direcionado com as seguintes características: 1 vértice com grau de entrada=3 e grau de saída = 2, 1 vértice com grau de entrada=2 e grau de saída = 0, 1 vértice com grau de entrada=0 e grau de saída = 1.

import networkx as nx

G = nx.DiGraph()

G.add_node(3)
G.add_node(2)
G.add_node(1)
G.add_node(0)

nx.draw(G, with_labels = True)
```
